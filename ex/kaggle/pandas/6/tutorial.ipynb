```python
# Setup /1: only pandas is used here
import pandas as pd
```

```python
# Setup /2: generate the data frame used for examples

reviews = pd.DataFrame({
    'country': ['Italy', 'Portugal', 'US', 'Italy', 'Canada', 'France', 'Italy','Australia', 'New Zealand'],
    'description': ['A', 'B tropical', 'C', 'D', 'E', 'F', 'E fruity', 'H', 'I'],
    'price': [12, 41, 34, 18, 27, 32, 43, 22, 12], 
    'winery': ['A 21', 'B 34', 'C 54', 'D 21', pd.NA, 'F 43', 'G 44', 'H 11', 'I 32'],
    'points': [85, 88, 87, 92, 81, 95, 84, 97, 80],
    'variety': ['a', 'b', 'a', 'b', 'a', 'b', 'a', 'b', 'a'],
    'title': ['a a', 'b b', 'c c', 'd d', 'e e', 'f f', 'g g', 'h h', 'i i'],
    'province': ['A', 'AP', 'AUSA', 'B', 'A', 'A', 'A','A', 'ANZ'],
    'taster_name': ['a', 'b', 'c', 'b', 'a', 'c', 'a', 'b', 'c'],
    'region_1': ['a', pd.NA, 'c', 'b', pd.NA, pd.NA, pd.NA, 'b', 'c'],
})

reviews.head(10)
```

# Renaming

rename() - change index/column names 

```python
# rename a column
reviews.rename(columns={'points': 'score'})
```

```python
# rename index by dict, key is the old name, value the new one
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

```python
# rename index (axis rows) and the axis columns
reviews.rename_axis("wines", axis='rows').rename_axis("fields", axis='columns')
```

# Combining

pd.concat() - simplest
pd.join()
pd.merge() - most complex

```python
# concat - same column names
more_reviews = pd.DataFrame({
    'country': ['France'],
    'description': ['F'],
    'price': [34], 
    'winery': ['A 21'],
    'points': [82],
    'variety': ['b'],
    'title': ['a a'],
    'province': ['A'],
    'taster_name': ['a'],
    'region_1': ['a'],
})

pd.concat([reviews, more_reviews])
```

```python
# join - same indices
left = reviews.set_index(['country', 'province'])
right = more_reviews.set_index(['country', 'province'])

left.join(right, lsuffix='_1', rsuffix='_2')
```
